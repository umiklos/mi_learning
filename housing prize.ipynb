```python
for i in range(1,17):
    rfe_i = RFE(lin_reg, i)
    rfe_i = rfe_i.fit(train_set, label_train)
    predictions_rfe_i = rfe_i.predict(train_set)
    lin_mse_rfe_i = mean_squared_error(label_train, predictions_rfe_i)
    lin_rmse_rfe_i = np.sqrt(lin_mse_rfe_i)
    print(i," ",lin_rmse_rfe_i)
```
